## Number of grains of rice on the 100th day

```python
def compute_sorori_shinzaemon(days):
#     total_grains = 1*((2**days)-1)
#     return total_grains
    day_grains = []
    total_grains = []
    for day in range(days):
        day_grains.append(2**day)
        total_grains.append(sum(day_grains))
    return day_grains,total_grains
list_n_grains, list_total_grains = compute_sorori_shinzaemon(100)
import matplotlib.pyplot as plt
%matplotlib inline
plt.title("Grain after 100 days")
plt.xlabel("Days")
plt.ylabel("Rice")
plt.plot(list_n_grains,label="Grains received each days")
plt.plot(list_total_grains,label="Total number of grains each days")
plt.legend() 
plt.show()
```
