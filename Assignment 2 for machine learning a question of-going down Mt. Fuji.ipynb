## [Problem 2]: Create a function to calculate the slope of a point

```python
def gradient(fuji):
    gradient = (fuji[1:,3]-fuji[0:-1,3])/(fuji[1:,0]-fuji[0:-1,0])
    return gradient
gradient =gradient(fuji)
gradient
```

## [Problem 3]: Create a function to calculate the destination

```python
def destination(current_location,gradient, a = 0.2):
    
    return round(current_location - a*gradient)
```

## Problem 4: Creating a function to go down the mountain

```python
%matplotlib inline
plt.xlabel("Position")
plt.ylabel("Elevation")
plt.title("Mt.Fuji")
plt.plot(fuji[:,0],fuji[:,3])
plt.scatter(136,fuji[136,3],c="Green",label="Current location for Point 136")
plt.legend()
```

```python
def go_down(first_point ,alpha = 0.3):

    dest = []
    dest.append(fuji[first_point,3])
    while first_point >= 0:
        dest.append(destination(dest[-1],gradient[first_point],alpha))
        first_point -= 1
    return dest
first_point = 136
down_road = go_down(136)
down_road
```

## [Problem 5]: Visualization of the descent process

```python
%matplotlib inline
plt.xlabel("X")
plt.ylabel("Elevation")
plt.title("Fuji")
plt.plot(range(len(down_road)),down_road)
```
