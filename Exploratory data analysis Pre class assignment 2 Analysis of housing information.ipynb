## [Problem 4] Dealing with missing values

```python
import numpy as np
from quilt.data.ResidentMario import missingno_data
collisions = missingno_data.nyc_collision_factors()
data = data.replace("NA", np.nan)
```

```python
nullset = data.isnull().sum()[data.isnull().sum()>0]
nullset
```

```python
percent_missing = data.isnull().sum() * 100 / len(data)
percent_missing = percent_missing[percent_missing>0]
percent_missing
missing_value_df = pd.DataFrame({'Total':nullset,
                                 'percent_missing': percent_missing})
missing_value_df
```

```python
missing_value_df.index
data = data.drop(columns = missing_value_df.index)
data
```

## [Problem 6]: Confirming distribution

```python
import seaborn as sns, numpy as np
sns.set_theme();
ax = sns.distplot(data.loc[:,"SalePrice"])
print("Skewness: " ,data.loc[:,"SalePrice"].skew())
print("Kurtosis: " ,data.loc[:,"SalePrice"].kurt())
```

```python
import numpy as np
nor_data = np.log(data.loc[:,"SalePrice"])
nor_data
```

```python
sns.set_theme();
ax = sns.distplot(nor_data)
print("Skewness: " ,nor_data.skew())
print("Kurtosis: " ,nor_data.kurt())
```
