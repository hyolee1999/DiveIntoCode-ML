```python
nullset = data.isnull().sum()[data.isnull().sum()>0]
nullset
```

```python
percent_missing = data.isnull().sum() * 100 / len(data)
percent_missing = percent_missing[percent_missing>0]
percent_missing
missing_value_df = pd.DataFrame({'Total':nullset,
                                 'percentage of missing': percent_missing})
missing_value_df
```

```python
import matplotlib.pyplot as plt
sizes = data["TARGET"].value_counts()
labels = 'different situations', 'client who have difficulties'

# plt.title("Distribution of Target Variable")
# plt.pie(label_counts, labels=labels, autopct='%.1f%%', startangle=90)
# plt.show()

fig1, ax1 = plt.subplots()
ax1.pie(sizes, labels=labels, autopct='%1.1f%%',
        shadow=True, startangle=90)
ax1.axis('equal')

plt.show()
```
