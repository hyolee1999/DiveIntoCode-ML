## Create a file for learning

```python
import pandas as pd
data=pd.read_csv('/content/drive/MyDrive/ObjectDetection/annotation.txt', names=["Path", "X1","X2","X3","X4","Name"])
```

```python
data
```

```python
data[data.columns[-1]]
```

```python
from sklearn import preprocessing
le = preprocessing.LabelEncoder()
data[data.columns[-1]]=le.fit_transform(data[data.columns[-1]])
```

```python
data
```

```python

f = open('/content/drive/MyDrive/ObjectDetection/annotation.txt','r')
```

```python
file = []

for i , line in enumerate(f.readlines()):
    line = "/content/drive/MyDrive/ObjectDetection/"+line.replace("jpg,","jpg ")
    line = line.split(",")
    line[-1] = data["Name"][i]
    line = ",".join(str(x)for x  in line)
    file.append(line)

f.close()

```

```python
f = open("/content/drive/MyDrive/keras-yolo3/annotation.txt","a")
for line in file:
  f.write(line+ "\n")
f.close()
```

```python
f = open("/content/drive/MyDrive/keras-yolo3/label.txt","a")
for line in list(le.classes_):
  f.write(line+ "\n")
f.close()
```

## Code reading

- Residual networks which help the activations propagate through deeper layers without gradient diminishing and can be found in model.py and  is implemented in yolo_body function

- Multi-scale Detector which is used to ensure that the small objects are also being detected unlike in YOLO v2 and can be found in model.py and  is implemented in yolo_eval function
