## Explain what is  different between loc and iloc

<li> <b> loc </b> is a label based data selecting method which means that we have to pass the name of the row or column which we want to select.
<li> <b> iloc </b> is a indexed based selecting method which means that we have to pass integer index in the method to select specific row/column.

## [Problem 6]: Creating a diagram

```python
import matplotlib.pyplot as plt
```

```python
from matplotlib import pyplot as plt
labels = 'Setosa', 'Virgicolor', 'Virginica'
sizes = [df[df["Species"] == 0]["Species"].count(), df[df["Species"] == 1]["Species"].count(), df[df["Species"] == 2]["Species"].count()]
explode = (0, 0, 0)  

fig1, ax1 = plt.subplots()
ax1.pie(sizes, explode=explode, labels=labels, autopct='%1.1f%%',
        shadow=True, startangle=90)
ax1.axis('equal')

plt.show()
```

```python
labels = 'sl', 'sw', 'pl','pw'

fig, (ax1, ax2,ax3) = plt.subplots(nrows=1, ncols=3, figsize=(9, 4))

# rectangular box plot
bplot1 = ax1.boxplot(df[df["Species"] == 0].iloc[:,0:4],vert=True,patch_artist=True,labels =labels)  # will be used to label x-ticks
ax1.set_title('Iris setosa')

bplot2 = ax2.boxplot(df[df["Species"] == 1].iloc[:,0:4],
                     vert=True,  # vertical box alignment
                     patch_artist=True,  # fill with color
                     labels=labels)  # will be used to label x-ticks
ax2.set_title('Iris virgicolor')

bplot3 = ax3.boxplot(df[df["Species"] == 2].iloc[:,0:4],
                     vert=True,  # vertical box alignment
                     patch_artist=True,  # fill with color
                     labels=labels)  # will be used to label x-ticks
ax3.set_title('Iris virginica')
# fill with colors
colors = ['pink', 'lightblue', 'lightgreen']
for bplot in (bplot1, bplot2, bplot3):
    for patch, color in zip(bplot['boxes'], colors):
        patch.set_facecolor(color)
   

# adding horizontal grid lines
for ax in [ax1, ax2, ax3]:
    ax.yaxis.grid(True)
    ax.set_xlabel('Widths and Lengths')
    ax.set_ylabel('feature  values')

plt.show()
```

```python
fig, axs = plt.subplots(nrows=1, ncols=3, figsize=(12, 6))

axs[0].violinplot(df[df["Species"] == 0].iloc[:,0:4], showmeans=True, showextrema=True, showmedians=True)
axs[0].set_title('Iris setosa')
# ax[0].set_xlabel('Four separate samples')
# ax[0].set_ylabel('Observed values')

axs[1].violinplot(df[df["Species"] == 1].iloc[:,0:4],showmeans=True, showextrema=True, showmedians=True)
axs[1].set_title('Iris virgicolor')

axs[2].violinplot(df[df["Species"] == 2].iloc[:,0:4],showmeans=True, showextrema=True, showmedians=True)
axs[2].set_title('Iris virginica')
for ax in axs.flat:
    ax.set_xticklabels(["","sl","sw","pl","pw"])
    ax.set_ylabel('Observed values')

fig.suptitle("Violin Plotting Examples")
fig.subplots_adjust(hspace=0.4)
plt.show()
```

# Explain how  plots and violin differ and what advantages each has.

Both are used to visualize numerical data. 

Box plots show data points outside 1.5 * the inter-quartile range as outliers above or below the whiskers whereas Violin plots show the whole range of the data.
